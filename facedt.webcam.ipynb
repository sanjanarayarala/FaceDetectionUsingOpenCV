```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('group.jpg') #Reads the image in BGR Format
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img) # Print image in the RGB format
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces = face_cascade.detectMultiScale(img,1.1)
```

```python
plt.imshow(img)
plt.show()
```

```python
# Create a Camera Object
camera = cv2.VideoCapture(0)

# Read an image
b,img = camera.read()
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)

# Close the Camera
camera.release()
```

```python
camera = cv2.VideoCapture(0)

while True:
    b,img = camera.read()
    if b==False:
        continue
    # Find out faces--------------------------
    faces = face_cascade.detectMultiScale(img,1.1)
    
    for f in faces:
        x,y,w,h = f
        green = (0,255,0)
        cv2.rectangle(img,(x,y),(x+w,y+h),green,5)
    #------------------------------------------
    cv2.imshow("Image ",img)
    key = cv2.waitKey(1)&0xFF
    if key==ord('q'):
        break
    
camera.release()   
cv2.destroyAllWindows()
```
